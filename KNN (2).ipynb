```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# 1. Load the Iris dataset
iris = load_iris()
X = iris.data  # Features
y = iris.target  # Labels

# 2. Split into training and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 3. Create the KNN classifier (k=3)
knn = KNeighborsClassifier(n_neighbors=3)

# 4. Train the classifier
knn.fit(X_train, y_train)

# 5. Predict on test set
y_pred = knn.predict(X_test)

# 6. Calculate accuracy
print("Accuracy:", accuracy_score(y_test, y_pred))

# 7. Print correct and wrong predictions
print("\nCorrect Predictions:")
for i in range(len(y_test)):
    if y_test[i] == y_pred[i]:
        print(f"Index: {i}, Predicted: {iris.target_names[y_pred[i]]}, Actual: {iris.target_names[y_test[i]]}")

print("\nWrong Predictions:")
for i in range(len(y_test)):
    if y_test[i] != y_pred[i]:
        print(f"Index: {i}, Predicted: {iris.target_names[y_pred[i]]}, Actual: {iris.target_names[y_test[i]]}")
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# 1. Load the Iris dataset
iris = load_iris()
X = iris.data  # Features
y = iris.target  # Labels

# 2. Split into training and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 3. Create the KNN classifier (initial k=3)
knn = KNeighborsClassifier(n_neighbors=3)

# 4. Train the classifier
knn.fit(X_train, y_train)

# 5. Predict on test set
y_pred = knn.predict(X_test)

# 6. Calculate accuracy
print("Accuracy (for k=3):", accuracy_score(y_test, y_pred))

# 7. Print correct and wrong predictions
print("\n‚úÖ Correct Predictions:")
for i in range(len(y_test)):
    if y_test[i] == y_pred[i]:
        print(f"Index: {i}, Predicted: {iris.target_names[y_pred[i]]}, Actual: {iris.target_names[y_test[i]]}")

print("\n‚ùå Wrong Predictions:")
for i in range(len(y_test)):
    if y_test[i] != y_pred[i]:
        print(f"Index: {i}, Predicted: {iris.target_names[y_pred[i]]}, Actual: {iris.target_names[y_test[i]]}")

# 8. Confusion Matrix for k=3
cm = confusion_matrix(y_test, y_pred)
print("\nüìå Confusion Matrix (k=3):")
print(cm)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=iris.target_names)
disp.plot(cmap='Blues')
plt.title(f"Confusion Matrix (k={3})")
plt.show()

# 9. Classification Report for k=3
print("\nüìë Classification Report (k=3):")
print(classification_report(y_test, y_pred, target_names=iris.target_names))

# 10. Test on different k values (from 1 to 10)
print("\nüìä Accuracy for different k values:")
for k in range(1, 11):
    knn_k = KNeighborsClassifier(n_neighbors=k)
    knn_k.fit(X_train, y_train)
    y_pred_k = knn_k.predict(X_test)
    acc = accuracy_score(y_test, y_pred_k)
    print(f"k = {k}: Accuracy = {acc:.2f}")

# 11. Hardcoded sample input prediction
print("\nüîç Test on hardcoded input:")

# Example: Hardcoded sample input (sepal length, sepal width, petal length, petal width)
sample = [5.1, 3.5, 1.4, 0.2]  # These are the features of the first sample in the Iris dataset
prediction = knn.predict([sample])
print(f"Predicted Iris species for sample {sample}: {iris.target_names[prediction[0]]}")
```

```python
# Step 1: Calculate Euclidean Distance
def euclidean_distance(point1, point2):
    distance = 0
    for i in range(len(point1)):
        distance += (point1[i] - point2[i]) ** 2
    return distance ** 0.5

# Step 2: Get K Nearest Neighbors
def get_neighbors(data, test_point, k):
    distances = []
    for row in data:
        dist = euclidean_distance(test_point, row[:-1])
        distances.append((dist, row))
    distances.sort(key=lambda x: x[0])
    neighbors = [distances[i][1] for i in range(k)]
    return neighbors

# Step 3: Do Majority Voting
def predict_classification(data, test_point, k):
    neighbors = get_neighbors(data, test_point, k)
    votes = {}
    for neighbor in neighbors:
        label = neighbor[-1]
        votes[label] = votes.get(label, 0) + 1
    predicted = max(votes, key=votes.get)
    return predicted

# Sample data
dataset = [
    [1, 2, 'A'],
    [2, 3, 'A'],
    [3, 3, 'B'],
    [6, 5, 'B'],
    [7, 8, 'B'],
    [8, 8, 'A']
]

test_point = [5, 5]
k = 3

# Make prediction
predicted_label = predict_classification(dataset, test_point, k)
print(f"Predicted class for point {test_point} is: {predicted_label}")
```

```python
from sklearn.datasets import load_iris
import random
import matplotlib.pyplot as plt

# Load dataset
iris = load_iris()
data = iris.data
labels = iris.target
target_names = iris.target_names

# Combine features and labels
dataset = list(zip(data, labels))

# Shuffle and split (80% train, 20% test)
random.seed(42)
random.shuffle(dataset)
split_index = int(0.8 * len(dataset))
train_data = dataset[:split_index]
test_data = dataset[split_index:]

# Euclidean distance function
def euclidean_distance(p1, p2):
    return sum((a - b) ** 2 for a, b in zip(p1, p2)) ** 0.5

# Get k nearest neighbors
def get_neighbors(train_data, test_point, k):
    distances = []
    for features, label in train_data:
        dist = euclidean_distance(features, test_point)
        distances.append((dist, label))
    distances.sort(key=lambda x: x[0])
    return [label for _, label in distances[:k]]

# Predict label based on neighbors
def predict_classification(train_data, test_point, k):
    neighbors = get_neighbors(train_data, test_point, k)
    vote_counts = {}
    for label in neighbors:
        vote_counts[label] = vote_counts.get(label, 0) + 1
    return max(vote_counts, key=vote_counts.get)

# Evaluate accuracy and confusion matrix
def evaluate_knn(train_data, test_data, k):
    correct = 0
    num_classes = len(set(labels))
    confusion = [[0] * num_classes for _ in range(num_classes)]

    for features, true_label in test_data:
        pred = predict_classification(train_data, features, k)
        confusion[true_label][pred] += 1
        if pred == true_label:
            correct += 1

    accuracy = correct / len(test_data)
    return accuracy, confusion

# Plot accuracy vs. k
def plot_accuracy_vs_k(train_data, test_data, max_k=20):
    ks = list(range(1, max_k + 1))
    accuracies = []

    for k in ks:
        acc, _ = evaluate_knn(train_data, test_data, k)
        accuracies.append(acc)

    plt.plot(ks, accuracies, marker='o')
    plt.title("KNN Accuracy vs K")
    plt.xlabel("K")
    plt.ylabel("Accuracy")
    plt.grid(True)
    plt.show()

# --- Main Execution ---
k = 5
accuracy, confusion = evaluate_knn(train_data, test_data, k)

print(f"Accuracy with k={k}: {accuracy:.2f}\n")

print("Confusion Matrix:")
for i, row in enumerate(confusion):
    print(f"{target_names[i]:<15}: {row}")

# Plot accuracy vs. k
plot_accuracy_vs_k(train_data, test_data, max_k=15)

# Predict on one hardcoded sample
hardcoded_sample = [5.9, 3.0, 5.1, 1.8]  # Should likely be Virginica
pred_label = predict_classification(train_data, hardcoded_sample, k)
print(f"\nPrediction for sample {hardcoded_sample} => {target_names[pred_label]}")
```

```python
# k-Nearest Neighbors (k-NN) ‚Äì Complete Guide

## üìå What is k-NN?

k-Nearest Neighbors (k-NN) is a **supervised learning algorithm** used for **classification and regression**.
It classifies a data point based on how its **neighbors are classified**.

---

## üí° Concept

- **Instance-based**: No model is trained. Data is stored and used at prediction time.
- **Lazy learning**: No training phase; predictions are made using stored data.
- **Majority voting**: For classification, return the most common class among the `k` nearest points.

---

## üî¢ Mathematical Formula

### Euclidean Distance (most commonly used):
\[
d(p, q) = \sqrt{(p_1 - q_1)^2 + (p_2 - q_2)^2 + \dots + (p_n - q_n)^2}
\]

Other distances:
- Manhattan: \( \sum |p_i - q_i| \)
- Minkowski: \( \left( \sum |p_i - q_i|^p \right)^{1/p} \)

---

## üõ†Ô∏è Steps of k-NN

1. Choose `k` (number of neighbors).
2. Compute distance from test point to all training data points.
3. Sort distances and find the `k` closest points.
4. Use majority voting to assign a class (for classification).
5. Return the predicted class.

---

## üîç Choosing the Value of k

- Small k (e.g., 1): Sensitive to noise.
- Large k: May include points from other classes.
- Prefer **odd k values** to avoid ties.
- Use **cross-validation** to select the optimal k.

---

## ‚úÖ Advantages

- Simple and intuitive.
- No training time needed.
- Supports multi-class classification.
- Easily adaptable to new data.

---

## ‚ùå Disadvantages

- High computation cost during prediction.
- Needs large memory (stores all training data).
- Poor performance with high-dimensional data.
- Sensitive to irrelevant/noisy features.
- Requires feature scaling.

---

## üìä Performance Metrics

### Confusion Matrix
A square matrix where:
- Rows = Actual class
- Columns = Predicted class

### Classification Report

| Metric     | Formula                              |
|------------|--------------------------------------|
| Precision  | TP / (TP + FP)                       |
| Recall     | TP / (TP + FN)                       |
| F1-score   | 2 √ó (Precision √ó Recall) / (Precision + Recall) |
| Accuracy   | (TP + TN) / (Total samples)          |

---

## üß† Common Viva Questions

**Q1. What is k-NN?**
A supervised algorithm that predicts a label based on the most common class among the k-nearest training points.

**Q2. Is it a lazy or eager learner?**
Lazy learner ‚Äî no model is trained.

**Q3. What is the role of distance in k-NN?**
Distance determines how "close" a point is to others; the closest ones are used for voting.

**Q4. Why is feature scaling needed?**
Because features with larger values can dominate the distance metric.

**Q5. What if there's a tie in voting?**
Use odd `k` values or break ties by choosing the closest sample among tied classes.

---

## üìÅ Use Cases

- Image classification
- Text classification
- Anomaly detection
- Recommender systems
- Customer behavior analysis

---

## üß™ Classification vs Regression in k-NN

| Task         | Method                                  |
|--------------|------------------------------------------|
| Classification | Majority vote from k neighbors        |
| Regression     | Average of k neighbors' output values |

---

## üìò Keywords to Remember

- k-nearest neighbors
- lazy learner
- Euclidean distance
- majority voting
- feature scaling
- classification report
- confusion matrix

---
```
